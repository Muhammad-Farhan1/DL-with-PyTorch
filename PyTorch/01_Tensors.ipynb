**Tensors in PyTorch** are multi-dimensional arrays that are **similar to NumPy arrays** but with additional capabilities for GPU acceleration. They are the fundamental building blocks for creating and training neural networks in PyTorch. Tensors can be created in various ways, including from lists, NumPy arrays, or by using built-in functions that generate tensors with specific properties (like random values, zeros, or ones).

```python
# Check PyTorch installation and CUDA availability
import torch
print(torch.__version__)
print("CUDA available:", torch.cuda.is_available())
```

# Creating tensors in PyTorch

```python
# Empty Tensor 
a = torch.empty(3,4)
print(a)
type(a)
```

```python
# Ones Tensor
b = torch.ones(2,3)
print(b)
```

```python
# Zeros Tensor
c = torch.zeros(4,5)
print(c)
```

```python
# Radom Tensor
d = torch.rand(2,2)
print(d)
```

```python
# Seed for reproducibility
torch.manual_seed(42)
# Random Tensor with seed
e = torch.rand(2,2)
print(e)
```

```python
# Identity Matrix
f = torch.eye(3)
print(f)
```

# Mathematical Operations on Tensors

# Scalar Operations

```python
a = torch.tensor([[1, 2, 3],[4, 5, 6]])
print(a)
```

```python
# Addition
a + 10
# Subtraction
a - 2
# Multiplication
a * 3
# Division
a / 2
# Exponentiation
a ** 2
# Modulus
a % 2
```

# Element-wise Operations

```python
f = torch.tensor([[1.4, 2.4], [3.7, 4.0]])
g = torch.tensor([[5.0, 6.1], [7.2, 8.3]])
print(f)
print(g)
```

```python
# Addition
f + g
# Subtraction
f - g
# Multiplication
f * g
# Division
f / g
```

```python
# Absolute Value
torch.abs(f)
# Square Root
torch.sqrt(f)
# Exponential
torch.exp(f)
# round
torch.round(f)
# ceil
torch.ceil(f)
# floor
torch.floor(f)
# clamp
torch.clamp(f, min=2.0, max=3.5)
```

# Comparison Operations

```python
a = torch.randint(size=(2, 3), low=0, high=10)
b = torch.randint(size=(2, 3), low=1, high=10)
print(a)
print(b)
```

```python
# equal
a == b
# greater than
a > b
# less than
a < b
# not equal
a != b
# greater than or equal to
a >= b
# less than or equal to
a <= b
```

# Copying a tensor

```python
j = torch.randint(size=(2, 3), low=0, high=10)
j
```

```python
# copying a tensor
k = j.clone()
k
```

```python
id(k) == id(j)
```
